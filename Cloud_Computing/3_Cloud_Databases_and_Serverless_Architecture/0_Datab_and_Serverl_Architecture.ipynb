# Cloud Databases and Serverless Architecture

1. [Serverless Computing](https://www.youtube.com/watch?v=bsVyjGpWJa4)

Serverless computing is a managed compute resource that provides backend services without the need for you to manage any infrastructure.

* Focus on development, not infrastructure
* Reduce the need for DevOps expertise
* Cost-effective - pay only for fractions of a second
* Infinite scalability

**When Serverless Is Not the Best Solution**

* Serverless is designed for short interactions and requests. If your server takes a long time to process a request or it needs a lot of memory, then serverless might not be a good solution.
* Serverless functions do not run constantly. You cannot install and run services such as real webserver with serverless computing.
